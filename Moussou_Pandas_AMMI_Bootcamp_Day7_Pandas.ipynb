### Groupby

```
import pandas as pd
# Create dataframe
data = {'Company':['GOOG','GOOG','MSFT','MSFT','FB','FB'],
       'Person':['Ahmed','Kobby','Aissa','Modou','Nando','Armand'],
       'Result':[200, 120, 340,124,243,350]}
```

```
df6 = pd.DataFrame(data)
df6
```

Now you can use the .groupby() method to group rows together based off of a column name. For instance let's group based off of Company. This will create a DataFrameGroupBy object:

```
df6.groupby('Company')
```

```
comp_mean = df6.groupby("Company").mean()
comp_mean
```

```
comp_std = df6.groupby("Company").std()
comp_std
```

```
by_comp = df6.groupby("Company")
```

```
by_comp.min()
```

```
by_comp.count()
```

```
by_comp.describe()
```

```
by_comp.describe().T
```

### Merging, Joining and Concatenating

There are 3 main ways of combining DataFrames together: Merging, Joining and Concatenating

```
df_a = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])
df_a
```

```
df_b = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7]) 
df_b
```

```
df_c = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8, 9, 10, 11])
df_c
```

Concatenation

```
pd.concat([df_a, df_b, df_c])
```

```
pd.concat([df_a, df_b, df_c], axis=1)
```

Merging

```
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```
left
```

```
right
```

```
pd.merge(left, right, on=['key1'])
```

```
pd.merge(left, right, on=['key1', 'key2'])
```

```
pd.merge(left, right, how='outer', on=['key1', 'key2'])
```

```
pd.merge(left, right, how='right', on=['key1', 'key2'])
```

```
pd.merge(left, right, how='left', on=['key1', 'key2'])
```

Joining

Joining is a convenient method for combining the columns of two potentially differently-indexed DataFrames into a single result DataFrame.

```
left = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
                     'B': ['B0', 'B1', 'B2']},
                      index=['K0', 'K1', 'K2'])

right = pd.DataFrame({'C': ['C0', 'C2', 'C3'],
                    'D': ['D0', 'D2', 'D3']},
                      index=['K0', 'K2', 'K3'])
```

```
left
```

```
right
```

```
left.join(right)
```

```
left.join(right, how='outer')
```
